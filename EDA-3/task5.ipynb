```python
import category_encoders as ce
```

```python
import pandas as pd
# инициализируем информацию об одежде
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing = pd.DataFrame(clothing_list, columns = ['size',  'type'])
clothing
```

```python
of = ce.OrdinalEncoder()
data_bin = of.fit_transform(clothing['size'])
clothing1 = pd.concat([clothing, data_bin], axis=1)
clothing1
```

```python
ohe = ce.OneHotEncoder(cols=['type'], use_cat_names=True)
type_bin = ohe.fit_transform(clothing['type'])
clothing2 = pd.concat([clothing1, type_bin], axis=1)

clothing2
```

```python
clothing3 = pd.get_dummies(clothing, columns=['type'] )
clothing3
```
